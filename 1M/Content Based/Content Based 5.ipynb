```python
import pandas as pd
import numpy as np

ratings = pd.read_csv('D:\\Escritorio\\UM\\Tesis\\ML 1M\\ratings.dat', sep='::',engine='python',
                      names=['user', 'item', 'rating', 'timestamp'])

ratings = ratings.drop(['timestamp'],axis=1)

rating = ratings.copy()

ratings.head(5)
```

# Boolean ratings

```python
boolean = ratings.copy()
boolean['rating'] = boolean['rating'].replace([1,2,3,4,5],1)
boolean.head()
```

# Genres Matrix

```python
genres = pd.read_csv('D:\\Escritorio\\UM\\Tesis\\ML 1M\\movies.dat', sep='::',engine='python',
                     names=['movieid', 'title', 'genre'])
idx = np.intersect1d(genres.movieid.values, rating.item.values)
genres = genres[genres['movieid'].isin(idx)]
genres = genres.join(genres.pop('genre').str.get_dummies())
genres = genres.drop(['title'],axis=1)
genres = genres.set_index('movieid')
genres = genres.divide(((((genres.T)**2).sum())**(1/2)),axis=0)
genres
```

# Pivot Ratings Matrix

```python
ratings = ratings.pivot(index = 'user',columns='item',values ='rating')
OldMin = 1
OldMax = 5
NewMin = -1
NewMax = 1
ratings = (((ratings - OldMin) * (NewMax - NewMin)) / (OldMax - OldMin)) + NewMin
ratings = ratings.fillna(0)
ratings
```

# Pivot Boolean matrix

```python
boolean = boolean.pivot(index = 'user',columns='item',values ='rating')
boolean = boolean.fillna(0)
boolean
```

# Profile matrix

```python
np.seterr(divide='ignore', invalid='ignore')
dot = boolean.values.dot(genres.values)
dot = pd.DataFrame(dot)
profile = pd.DataFrame((ratings).values.dot(genres.values), columns=genres.columns, index=(ratings).index)
profile = profile.values/(dot.values)
profile = pd.DataFrame(profile)
profile.columns = genres.columns
profile = profile.fillna(0)
profile
```

# Compute Predictions 

```python
count = genres.sum(axis=1)
count = (pd.DataFrame(count))
prediction = pd.DataFrame((genres.values.dot((profile.T).values))/count.values, columns=(profile.T).columns, index=genres.index)
prediction = prediction.T
prediction.columns = ratings.columns
prediction.index = ratings.index
prediction
```

# Reset to 1-5 star scale

```python
OldMin = -1
OldMax = 1
NewMin = 1
NewMax = 5
prediction = (((prediction - OldMin) * (NewMax - NewMin)) / (OldMax - OldMin)) + NewMin
prediction
```

```python
ratings = pd.read_csv('D:\\Escritorio\\UM\\Tesis\\ML 1M\\ratings.dat', sep='::',engine='python',
                      names=['user', 'item', 'rating', 'timestamp'])
ratings = ratings.drop(['timestamp'],axis=1)
ratings = ratings.pivot(index = 'user',columns='item',values ='rating')
ratings
```

# MAE

```python
%%time
MAE = abs(ratings-prediction)
MAE.mean(axis=1).mean()
```

# RMSE

```python
%%time
RMSE = ((((MAE**2).mean(axis=1)))**(1/2)).mean()
RMSE
```

```python
predicted = []
users = []
for i in range(0,prediction.shape[0]):
    predicted.append(prediction.iloc[i,:].nlargest(100))
    users.append([i+1]*100)
predicted = pd.concat(predicted, ignore_index=False)
predicted = pd.DataFrame(predicted)
predicted.columns = ['score']
user = pd.DataFrame(users)
user = pd.DataFrame(user.values.flatten())
user.index = predicted.index
predicted['user'] = user
predicted.reset_index(level=0, inplace=True)
predicted
```

```python
from lenskit import batch, topn, util
from lenskit import crossfold as xf
from lenskit import topn
```

# nDCG

```python
def model_ndcg():
    test_data = []
    for train, test in xf.partition_users(rating[['user', 'item', 'rating']], 5, xf.SampleFrac(0.2)):
        test_data.append(test)
    test_data = pd.concat(test_data, ignore_index=True)

    rla = topn.RecListAnalysis()
    rla.add_metric(topn.ndcg)
    results = rla.compute(predicted, test_data)

    return results.ndcg.mean()
```

```python
%%time
array = []
for i in range(1,100,1):
    array.append(model_ndcg())
```

```python
metric = pd.DataFrame(array)
metric.columns = ['nDCG']
print(metric.mean())
metric.sort_values(by=['nDCG'],ascending=False).head(5)
```

```python
metric = metric.sort_values(by=['nDCG'],ascending=False)
metric.head(10)
```

# F1, Precision, Recall

```python
def model_f1():
    test_data = []
    for train, test in xf.partition_users(rating[['user', 'item', 'rating']], 5, xf.SampleFrac(0.2)):
        test_data.append(test)
    test_data = pd.concat(test_data, ignore_index=True)
    rla1 = topn.RecListAnalysis()
    rla2 = topn.RecListAnalysis()
    rla1.add_metric(topn.precision)
    rla2.add_metric(topn.recall)
    prec = rla1.compute(predicted, test_data)
    rec = rla2.compute(predicted, test_data)
    F1 = ((prec['precision'].values*rec['recall'].values*2)/(prec['precision'].values+rec['recall'].values))
    F1 = np.nan_to_num(F1)
    return [F1.mean(),prec['precision'].mean(),rec['recall'].mean()]
```

```python
%%time
array = []
for i in range(1,100,1):
    array.append(model_f1())
```

```python
metric = pd.DataFrame(array)
metric.columns = ['F1','Precision','Recall']

metric = metric.sort_values(by=['F1'],ascending=False)
print(metric.head(10).mean())
```
