# 1) Item-Genre matrix

# Import data

```python
movies = pd.read_csv('D:\\Escritorio\\UM\\Tesis\\ML 20M\\ml-20m\\movies.csv', sep=',',header='infer')
movies.columns = ['item','title','genres']
movies.head(10)
```

# Turn genres strings into lists

```python
movies['genres'] = [x.split('|') for x in movies['genres']]
```

# Explode lists

```python
movies = movies.explode('genres')
movies.drop(['title'],axis=1)
movies['value'] = 1
movies.head(10)
```

# Pivot

```python
movies = movies.pivot(index = 'item',columns='genres',values ='value')
movies = movies.fillna(0)
movies
```

# Normalize

```python
movies = movies.divide(((((movies.T)**2).sum())**(1/2)),axis=0)
movies
```

```python
movies.to_csv('itemgenre.csv', index=False)
```
